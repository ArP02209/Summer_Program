```python
import cv2,socket,numpy
```

```python
s=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
```

```python
#Connecting to internal webcam
cap=cv2.VideoCapture(0)
```

```python
IP="192.168.54.151"
port_no= 2222
```

```python
#Binding IP and port no. : listening 1234 ...
s.bind((IP,port_no))

s.listen(5)
```

```python
#Accepting client request and maintain state
c,addr=s.accept()
```

```python
while True:
    
    print("Receiving ...")
    reply = c.recv(10000000)
    arr = numpy.frombuffer(reply, numpy.uint8,count=-1)
    client_photo = cv2.imdecode(arr,1)
    if type(client_photo) is type(None):
        pass
    else:
        cv2.imshow("CLIENT",client_photo)
        if cv2.waitKey(10)==13:
            break
    
    ret , photo=cap.read()
    photo=cv2.imencode(".jpg",photo)[1].tobytes()
    c.sendall(photo)
    print("Sending ...")

cv2.destroyAllWindows()
cap.release()
```
