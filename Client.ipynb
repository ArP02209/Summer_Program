```python
import socket,numpy,cv2
```

```python
#Creating Socket
s=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
```

```python
#Connecting to external webcam
cap=cv2.VideoCapture(1)
```

```python
#IP, Port no. of server
IP="192.168.54.151"
Port_no=2222
```

```python
#Creates a session with server
s.connect((IP,Port_no))
```

```python
while True:
    print("Sending ...")
    ret , photo=cap.read()
    photo=cv2.imencode(".jpg",photo)[1].tobytes()
    s.sendall(photo)
    
    reply = s.recv(10000000)
    arr = numpy.frombuffer(reply, numpy.uint8,count=-1)
    server_photo = cv2.imdecode(arr,1)
    if type(server_photo) is type(None):
        pass
    else:
        cv2.imshow("SERVER",server_photo)
        if cv2.waitKey(10)==13:
            break
    print("Receiving")

cv2.destroyAllWindows()
cap.release()
```
